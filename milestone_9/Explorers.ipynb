```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("database.csv")  
df.head()
```

2) Add a Salary column, which is generated randomly from -10000 to 40000

```python
df['salary'] = np.random.randint(-10000, 40000, size=len(df))
df.head()
```

3) Add a 'Correct salary' column with True value if Salary >= 0, else False value

```python
df['Correct salary'] = df['salary'] >= 0
df.head()
```

4) If the Salary column in the data frame is < 0, change to 0

```python
df.loc[df['salary'] < 0, 'salary'] = 0
df.head()
```

5) Save the new data frame to file database_with_salary.csv

```python
df.to_csv('database_with_salary.csv')
```

6) For each "Department" sum the "Salary" and create a bar chart with all departments and the total salary

```python
# Групування даних за департаментами
groupped_df = df.groupby('department')
result_table = groupped_df[['salary']].agg('sum').reset_index()
result_table
```

```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors

plt.figure(figsize=(8, 6))

norm = mcolors.Normalize(vmin=result_table['salary'].min(), vmax=result_table['salary'].max())
colors = cm.Greens(norm(result_table['salary']))

bars = plt.bar(result_table['department'], result_table['salary'], color=colors, edgecolor='black')
plt.xlabel('Department')
plt.ylabel('Total Salary, $')
plt.title('Розподіл зарплатного фонду за департаментами')
plt.xticks(rotation=45) 
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1000, f'{yval:,.0f}', ha='center', va='bottom', color='black')

plt.legend(['Total Salary per Department'], loc='upper right')
plt.show()


```
